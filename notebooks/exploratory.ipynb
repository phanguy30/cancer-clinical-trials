```python
import pandas as pd
import matplotlib.pyplot as plt

trials = pd.read_csv("../data/processed/trials.csv")

status_counts = trials["overall_status"].value_counts()

plt.figure()
status_counts.plot(kind="bar")
plt.title("Clinical Trials by Overall Status")
plt.xlabel("Status")
plt.ylabel("Number of Trials")
plt.tight_layout()
plt.show()
```

```python
study_type_counts = trials["study_type"].value_counts()

plt.figure()
study_type_counts.plot(kind="bar")
plt.title("Trials by Study Type")
plt.xlabel("Study Type")
plt.ylabel("Number of Trials")
plt.tight_layout()
plt.show()
```

```python
conditions = pd.read_csv("../data/processed/conditions.csv")

top_conditions = conditions["condition"].value_counts().head(20)

plt.figure()
top_conditions.sort_values().plot(kind="barh")
plt.title("Top 20 Conditions in Cancer Trials")
plt.xlabel("Number of Trials")
plt.ylabel("Condition")
plt.tight_layout()
plt.show()
```

```python
sponsors = pd.read_csv("../data/processed/sponsors.csv")

top_sponsors = sponsors["lead_sponsor_name"].value_counts().head(15)

plt.figure()
top_sponsors.sort_values().plot(kind="barh")
plt.title("Top Sponsors by Number of Trials")
plt.xlabel("Number of Trials")
plt.ylabel("Sponsor")
plt.tight_layout()
plt.show()
```
